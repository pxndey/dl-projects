```python
scaler = StandardScaler()
```

```python
x = df.drop('Class',axis=1)
x
```

```python
y = df['Class']
y
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=101)
```

```python
s_xtrain = scaler.fit_transform(xtrain)
```

```python
s_xtest = scaler.transform(xtest)
```

```python
model = Sequential(
    [
        layers.Dense(8,activation='relu'),
        layers.Dense(4,activation='relu'),
        layers.Dense(1,activation='sigmoid'),
    ]
)
```

```python
model
```

```python
model.compile(optimizer=Adam(0.001),loss=tf.keras.losses.BinaryCrossentropy())
```

```python
model.fit(xtrain,ytrain,epochs=10)
```

```python
ypreds  = model.predict(xtest)
```

```python
ytest
```

```python
ypreds
```

```python
binary_ypreds =(ypreds>0.5).astype(int)
```

```python
binary_ypreds
```

```python
accuracy_score(ytest,binary_ypreds)
```

```python
f1_score(ytest,binary_ypreds)
```

```python
precision_score(ytest,binary_ypreds)
```

```python
recall_score(ytest,binary_ypreds)
```
